```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Processed.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
severe_accidents = df[df['Severity'] == 4]

# Count the occurrences of each unique 'County' value
severe_count_by_county = severe_accidents['County'].value_counts()

# Find the county with the most severe accidents
most_severe_county = severe_count_by_county.idxmax()
most_severe_accidents = severe_count_by_county.max()
```

```python
print(f"The county with the most severe accidents (Severity=4) is {most_severe_county} with {most_severe_accidents} severe accidents.")
```
